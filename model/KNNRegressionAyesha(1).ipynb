## KNN Regression Training using Mean Squared Error

Predicting the unamed columns 



```python
import pandas as pd
import numpy as np
from shutil import copyfile
import os
```

```python
src = "../data/bigtable.csv"
dst = "./bigtable.csv"
if os.path.exists(dst):
    os.remove(dst)
copyfile(src, dst)
```

```python
df = pd.read_csv(dst)
df.tail()
```

```python
from pandas.plotting import scatter_matrix
#scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal='kde')
```

This will give the dimension of 2-d Array 

```python
from numpy import set_printoptions
df2=df.drop(["datetime", "type", "sensor", "station_id", "x", "y"], axis=1)
```

```python
df2 = df2.dropna()
```

```python
df2.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor

array = df2.values
print(len(array))
print(len(array[0]))

X = array[:,2:16]
Y = array[:,1]

neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(X, Y)

Y_target = neigh.predict(X)
```

```python
Y_target
```

```python
np.mean((Y_target - Y)**2)

# mse
#np.mean always computes an arithmetic mean, and has some additional options for input and output (e.g. what datatypes to use, where to place the result).
```
